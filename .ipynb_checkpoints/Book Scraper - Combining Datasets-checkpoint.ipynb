```python
import pandas as pd
df1 = pd.read_csv("Books.csv")
df2 = pd.read_csv("Data.csv")
```

```python
df1.head()
```

```python
df2.head()
```

```python
df = pd.DataFrame() # creates an empty dataframe
```

```python
df["TITLE"] = df1["TITLE"] # adding title column to the empty dataframe
df["CATEGORY"] = df2["CATEGORY"] # adding category column
df["PRICE (EXCL. TAX)"] = df2["PRICE (EXCL. TAX)"] # adding price info
df["PRICE (INCL. TAX)"] = df2["PRICE (INCL. TAX)"]
df["TAX"] = df2["TAX"]
df["RATING"] = df1["RATING"] # adding rating column
df["NUMBER OF REVIEWS"] = df2["NO. OF REVIEWS"] # adding no. of reviews
df["UPC"] = df2["UPC"] # adding upc
df["AVAILABILITY"] = df2["AVAILIABILITY"] # adding quantity
df["BOOK LINK"] = df1["BOOK LINK"] # adding book link
df["IMAGE LINK"] = df1["IMAGE LINK"] # adding image link
```

```python
df.to_csv("Books Info.csv", index = False)
```
