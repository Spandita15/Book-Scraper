# Scraping Books Data from Home - page

We see that inside this li tag we have an article tag inside which we have div which is the image container, and inside this we have an index.html site which has the details of the book along with the image. So we will scrape the data from one page and also the data of each page of index.html. Inside the div, we also have the image link and the alt tag has the title of the book. Out of div tag we have the p tag which has the star rating of the book. Inside the div tag we also have the anchor tag storing the link. We can also get the price and if its in stock or not using the respective tags.

```python
# lets see how many anchor tags we have in one book.
for sp in soup.find_all("li", class_="col-xs-6 col-sm-4 col-md-3 col-lg-3"):
    print(sp)
    print("--------")
    print(sp.find_all("a"))
    break
# we will see there are 2 anchor tags
```

```python
# in first anchor tag we have the image link and the index.html link and the second has the same index.html link and also the title 
# lets take both of them
for sp in soup.find_all("li", class_="col-xs-6 col-sm-4 col-md-3 col-lg-3"):
    print(sp)
    print("--------")
    link = sp.find_all("a")[-1].get("href")
    title = sp.find_all("a")[-1].get("title")
    break
```

```python
# lets do it for all the books
for sp in soup.find_all("li", class_="col-xs-6 col-sm-4 col-md-3 col-lg-3"):
    print("TITLE        : ",sp.find_all("a")[-1].get("title"))
    print("BOOK DETAILS : ", sp.find_all("a")[-1].get("href"))
    print("IMAGE LINK   : ", sp.find("img").get("src"))
    print("RATING       : ", sp.find("p").get("class")[-1])
    print("PRICE        : ", sp.find("p", class_="price_color").text[1:])
    print("AVAILABILITY : ", sp.find("p", class_="instock availability").text.strip())
    print("--------"* 10)
```

```python
# lets save this on our local system
data = []
for sp in soup.find_all("li", class_="col-xs-6 col-sm-4 col-md-3 col-lg-3"):
    title = sp.find_all("a")[-1].get("title")
    book = "https://books.toscrape.com/catalogue/" + sp.find_all("a")[-1].get("href")
    img_link = "https://books.toscrape.com" + sp.find("img").get("src")[2:]
    rating = sp.find("p").get("class")[-1]
    price = sp.find("p", class_="price_color").text[1:]
    avbl = sp.find("p", class_="instock availability").text.strip()
    data.append([title, rating, price, avbl, book, img_link])
```
